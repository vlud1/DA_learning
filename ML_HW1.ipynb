### Предсказать параметр medv из заданного датасета 

```python
import pandas as pd
df = pd.read_table('https://people.sc.fsu.edu/~jburkardt/datasets/boston_housing/boston_housing.txt', 
                    sep='[ ]{1,}', #один или более пробелов подряд
                    header = None,
                    names=["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "black", "lstat", "medv"],
                    engine='python')
df
```

```python
#проверяем на пропуски
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
random_state=69
```

```python
#разделяем данные для обучения и прогнозирования
x_train, x_test, y_train, y_test = train_test_split(df.drop('medv',axis=1), 
                                                    df['medv'], 
                                                    test_size=0.3, 
                                                    random_state=random_state)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
%%time
#обучаем модель по 70% данных (линейная регрессия)
medv_train_model = LinearRegression()
medv_train_model.fit(x_train, y_train)
```

```python
#по обученной модели делаем предсказания по тестовой выборке
pred = medv_train_model.predict(x_test)
```

```python
compare_predictions_table = x_test.copy()
compare_predictions_table['origin_medv'] = y_test
compare_predictions_table['pred_medv'] = pred
compare_predictions_table
```

```python
#сравним предсказанные и оригинальные значения "medv"
compare_predictions_table[['origin_medv', 'pred_medv']]
```

```python
#оценим прогноз по среднеквадратической ошибке
from sklearn.metrics import mean_squared_error
import numpy as np
print(f'Среднеквадратическая ошибка (RMSE) c random_state = {random_state} по прогнозным значениям составляет {np.round(mean_squared_error(y_test, pred), decimals=2)}')
```

```python
#испытаем метод 'Random Forest' для прогнозирования параметра 'medv'
from sklearn.ensemble import RandomForestRegressor
rf_medv_prediction_model = RandomForestRegressor()
rf_medv_prediction_model.fit(x_train, y_train)
```

```python
#прогноз по методу случайных деревьев
pred2 = rf_medv_prediction_model.predict(x_test)
```

```python
compare_predictions_table['pred2_medv'] = pred2
compare_predictions_table[['origin_medv', 'pred2_medv']]
```

```python
print(f'Среднеквадратическая ошибка (RMSE) c random_state = {random_state} по прогнозным значениям составляет {np.round(mean_squared_error(y_test, pred2), decimals=2)}')
```

```python
#дополнительно протестируем метод чрезвычайно рандомизированных деревьев
from sklearn.ensemble import ExtraTreesRegressor
etr_medv_prediction_model = ExtraTreesRegressor()
etr_medv_prediction_model.fit(x_train, y_train)
```

```python
#прогноз
pred3 = etr_medv_prediction_model.predict(x_test)
compare_predictions_table['pred3_medv'] = pred3
compare_predictions_table[['origin_medv', 'pred3_medv']]
```

```python
print(f'Среднеквадратическая ошибка (RMSE) c random_state = {random_state} по прогнозным значениям составляет {np.round(mean_squared_error(y_test, pred3), decimals=2)}')
```

### по итогу исследования выявлено, что при установленном параметре random_state более точно параметр medv предсказывают методы случайных деревьев и чрезвычайно рандомизированных деревьев
